```python
import os

# change these to your dataset paths
train_labels_dir = r"C:\Users\aadit\Downloads\BFD1\train\labels"
val_labels_dir   = r"C:\Users\aadit\Downloads\BFD1\valid\labels"

def count_labels(labels_dir):
    total_files = 0
    total_boxes = 0
    empty_files = 0

    for fname in os.listdir(labels_dir):
        if not fname.endswith(".txt"):
            continue
        total_files += 1
        path = os.path.join(labels_dir, fname)
        with open(path, "r") as f:
            lines = [ln.strip() for ln in f.readlines() if ln.strip()]
        if not lines:
            empty_files += 1
        total_boxes += len(lines)

    return {
        "label_files": total_files,
        "total_boxes": total_boxes,
        "empty_files": empty_files,
        "avg_boxes_per_image": total_boxes / total_files if total_files else 0
    }

train_stats = count_labels(train_labels_dir)
val_stats = count_labels(val_labels_dir)

print("TRAIN:", train_stats)
print("VALID:", val_stats)
```

```python
import os
import shutil
from pathlib import Path
import random
import cv2  # pip install opencv-python

# -----------------------------
# SOURCE PATHS
# -----------------------------
# Your existing BFD1 dataset (fractures)
BFD1_TRAIN_IMG = r"C:\Users\aadit\Downloads\BFD1\train\images"
BFD1_TRAIN_LAB = r"C:\Users\aadit\Downloads\BFD1\train\labels"

BFD1_VAL_IMG   = r"C:\Users\aadit\Downloads\BFD1\valid\images"
BFD1_VAL_LAB   = r"C:\Users\aadit\Downloads\BFD1\valid\labels"

BFD1_TEST_IMG  = r"C:\Users\aadit\Downloads\BFD1\test\images"
BFD1_TEST_LAB  = r"C:\Users\aadit\Downloads\BFD1\test\labels"

# FracAtlas paths
FRAC_ATLAS_FRACTURED_IMG = r"C:\Users\aadit\Downloads\FracAtlas\FracAtlas\images\Fractured"
FRAC_ATLAS_HEALTHY_IMG   = r"C:\Users\aadit\Downloads\FracAtlas\FracAtlas\images\Non_fractured"
FRAC_ATLAS_LABELS        = r"C:\Users\aadit\Downloads\FracAtlas\FracAtlas\Annotations\YOLO"

# -----------------------------
# DESTINATION (combined) PATHS
# -----------------------------
ROOT_COMBINED = r"C:\Final project 2\backend\datasets\Combined_Set"
IMG_TRAIN_DST = os.path.join(ROOT_COMBINED, "images", "train")
IMG_VAL_DST   = os.path.join(ROOT_COMBINED, "images", "valid")
IMG_TEST_DST  = os.path.join(ROOT_COMBINED, "images", "test")

LAB_TRAIN_DST = os.path.join(ROOT_COMBINED, "labels", "train")
LAB_VAL_DST   = os.path.join(ROOT_COMBINED, "labels", "valid")
LAB_TEST_DST  = os.path.join(ROOT_COMBINED, "labels", "test")

for d in [IMG_TRAIN_DST, IMG_VAL_DST, IMG_TEST_DST, LAB_TRAIN_DST, LAB_VAL_DST, LAB_TEST_DST]:
    os.makedirs(d, exist_ok=True)

# -----------------------------
# helper: CLAHE / histogram enhancement
# -----------------------------
def enhance_image(src_path, dst_path):
    img = cv2.imread(src_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        shutil.copy2(src_path, dst_path)
        return
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    enhanced = clahe.apply(img)
    cv2.imwrite(dst_path, enhanced)

# -----------------------------
# 1) COLLECT FRACTURED IMAGES (BFD1 train/val/test + FracAtlas fractured)
# -----------------------------
fractured_items = []

def collect_bfd1(img_dir, lab_dir, tag):
    out = []
    for f in Path(img_dir).glob("*.*"):
        label_name = f.with_suffix(".txt").name
        lab_path = Path(lab_dir) / label_name
        if lab_path.exists():
            out.append((tag, str(f), str(lab_path)))
    return out

fractured_items += collect_bfd1(BFD1_TRAIN_IMG, BFD1_TRAIN_LAB, "bfd1_train")
fractured_items += collect_bfd1(BFD1_VAL_IMG,   BFD1_VAL_LAB,   "bfd1_valid")
fractured_items += collect_bfd1(BFD1_TEST_IMG,  BFD1_TEST_LAB,  "bfd1_test")

# FracAtlas fractured
for f in Path(FRAC_ATLAS_FRACTURED_IMG).glob("*.*"):
    label_name = f.with_suffix(".txt").name
    label_path = Path(FRAC_ATLAS_LABELS) / label_name
    if label_path.exists():
        fractured_items.append(("fracatlas", str(f), str(label_path)))

print(f"Total fractured images found: {len(fractured_items)}")

# -----------------------------
# 2) COLLECT HEALTHY IMAGES
# -----------------------------
healthy_items = [str(f) for f in Path(FRAC_ATLAS_HEALTHY_IMG).glob("*.*")]
print(f"Total healthy images found: {len(healthy_items)}")

# -----------------------------
# 3) BALANCE (same number fractured vs healthy)
# -----------------------------
n_use = min(len(fractured_items), len(healthy_items))
random.shuffle(fractured_items)
random.shuffle(healthy_items)
fractured_items = fractured_items[:n_use]
healthy_items   = healthy_items[:n_use]
print(f"Using {n_use} fractured and {n_use} healthy images.")

# -----------------------------
# 4) SPLIT into train/val/test 80/10/10
# -----------------------------
def split_3(items, train_ratio=0.8, val_ratio=0.1):
    n = len(items)
    n_train = int(n * train_ratio)
    n_val = int(n * val_ratio)
    train = items[:n_train]
    val = items[n_train:n_train + n_val]
    test = items[n_train + n_val:]
    return train, val, test

frac_train, frac_val, frac_test = split_3(fractured_items, 0.8, 0.1)
heal_train, heal_val, heal_test = split_3(healthy_items, 0.8, 0.1)

# -----------------------------
# 5) SAVE FRACTURED (with labels)
# -----------------------------
def save_fractured(batch, img_dst_root, lab_dst_root, prefix="fracture"):
    for i, (src_type, img_path, lab_path) in enumerate(batch):
        base_name = f"{prefix}_{src_type}_{i:05d}.jpg"
        dst_img_path = os.path.join(img_dst_root, base_name)
        dst_lab_path = os.path.join(lab_dst_root, base_name.replace(".jpg", ".txt"))

        enhance_image(img_path, dst_img_path)
        shutil.copy2(lab_path, dst_lab_path)

# train / val / test for fractured
save_fractured(frac_train, IMG_TRAIN_DST, LAB_TRAIN_DST, "fracture")
save_fractured(frac_val,   IMG_VAL_DST,   LAB_VAL_DST,   "fracture")
save_fractured(frac_test,  IMG_TEST_DST,  LAB_TEST_DST,  "fracture")

# -----------------------------
# 6) SAVE HEALTHY (empty label files)
# -----------------------------
def save_healthy(batch, img_dst_root, lab_dst_root, prefix="healthy"):
    for i, img_path in enumerate(batch):
        base_name = f"{prefix}_{i:05d}.jpg"
        dst_img_path = os.path.join(img_dst_root, base_name)
        dst_lab_path = os.path.join(lab_dst_root, base_name.replace(".jpg", ".txt"))

        enhance_image(img_path, dst_img_path)
        with open(dst_lab_path, "w") as f:
            pass  # empty -> no objects

# train / val / test for healthy
save_healthy(heal_train, IMG_TRAIN_DST, LAB_TRAIN_DST, "healthy")
save_healthy(heal_val,   IMG_VAL_DST,   LAB_VAL_DST,   "healthy")
save_healthy(heal_test,  IMG_TEST_DST,  LAB_TEST_DST,  "healthy")

# -----------------------------
# 7) WRITE data.yaml
# -----------------------------
data_yaml_path = os.path.join(ROOT_COMBINED, "data.yaml")
with open(data_yaml_path, "w") as f:
    f.write(
        "path: C:/Final project 2/backend/datasets/Combined_Set\n"
        "train: images/train\n"
        "val: images/valid\n"
        "test: images/test\n"
        "nc: 2\n"
        "names: ['Fracture', 'Healthy']\n"
    )

print("Combined dataset created at:", ROOT_COMBINED)
```

```python
import os
from pathlib import Path
from ultralytics import YOLO
import numpy as np

# === paths ===
img_dir = Path(r"C:\Final project 2\backend\datasets\Combined_Set\images\test")
label_dir = Path(r"C:\Final project 2\backend\datasets\Combined_Set\labels\test")

detector = YOLO(r"C:\Final project 2\backend\runs_combined\yolo11m_combined_640_long\weights\best.pt")
classifier = YOLO(r"C:\Final project 2\backend\runs_cls\yolo11m_cls_fracture\weights\best.pt")

# === counters ===
tp = fp = tn = fn = 0

def is_fracture_gt(label_path):
    """Return True if label file has any boxes (fracture present)."""
    return os.path.isfile(label_path) and os.path.getsize(label_path) > 0

for img_path in img_dir.glob("*.*"):
    if not img_path.suffix.lower() in [".jpg", ".png", ".jpeg"]:
        continue

    # --- ground truth ---
    label_path = label_dir / (img_path.stem + ".txt")
    gt_fracture = is_fracture_gt(label_path)

    # --- classifier ---
    cls_res = classifier.predict(str(img_path), verbose=False)[0]
    cls_id = int(cls_res.probs.top1)
    cls_label = cls_res.names[cls_id].lower()
    cls_conf = float(cls_res.probs.top1conf)

    # --- detector ---
    det_res = detector.predict(str(img_path), conf=0.2, iou=0.55, max_det=300, verbose=False)[0]
    num_boxes = len(det_res.boxes)

    # --- decision logic ---
    if num_boxes > 0:
        pred_fracture = True
    elif cls_label == "fracture" and cls_conf > 0.7:
        pred_fracture = True  # "possible fracture"
    else:
        pred_fracture = False

    # --- compare ---
    if gt_fracture and pred_fracture:
        tp += 1
    elif not gt_fracture and pred_fracture:
        fp += 1
    elif not gt_fracture and not pred_fracture:
        tn += 1
    elif gt_fracture and not pred_fracture:
        fn += 1

# === metrics ===
precision = tp / (tp + fp + 1e-9)
recall = tp / (tp + fn + 1e-9)
f1 = 2 * precision * recall / (precision + recall + 1e-9)
acc = (tp + tn) / (tp + tn + fp + fn + 1e-9)

print("\n=== PIPELINE (Classifier + Detector) RESULTS ON TEST ===")
print(f"Total images: {tp+tn+fp+fn}")
print(f"TP: {tp}  FP: {fp}  TN: {tn}  FN: {fn}")
print(f"Accuracy:  {acc:.3f}")
print(f"Precision: {precision:.3f}")
print(f"Recall:    {recall:.3f}")
print(f"F1-score:  {f1:.3f}")
```

```python
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ultralytics import YOLO

# === paths ===
img_dir = Path(r"C:\Final project 2\backend\datasets\Combined_Set\images\test")
label_dir = Path(r"C:\Final project 2\backend\datasets\Combined_Set\labels\test")
save_dir = Path(r"C:\Final project 2\backend\notebooks\runs\detect\val_classifier_integration")
save_dir.mkdir(parents=True, exist_ok=True)

# === load models ===
detector = YOLO(r"C:\Final project 2\backend\runs_combined\yolo11m_combined_640_long\weights\best.pt")
classifier = YOLO(r"C:\Final project 2\backend\runs_cls\yolo11m_cls_fracture\weights\best.pt")

def is_fracture_gt(label_path: Path) -> bool:
    return label_path.exists() and label_path.stat().st_size > 0

# === containers ===
precisions, recalls, f1s, accs = [], [], [], []
steps = []
tp = fp = tn = fn = 0
y_true, y_pred = [], []

# === iterate images ===
all_imgs = sorted([p for p in img_dir.glob("*.*") if p.suffix.lower() in [".jpg", ".jpeg", ".png"]])

for idx, img_path in enumerate(all_imgs, start=1):
    label_path = label_dir / (img_path.stem + ".txt")
    gt_fracture = is_fracture_gt(label_path)

    # --- classifier ---
    cls_res = classifier.predict(str(img_path), verbose=False)[0]
    cls_id = int(cls_res.probs.top1)
    cls_label = cls_res.names[cls_id].lower()
    cls_conf = float(cls_res.probs.top1conf)

    # --- detector ---
    det_res = detector.predict(str(img_path), conf=0.2, iou=0.55, max_det=300, verbose=False)[0]
    has_boxes = len(det_res.boxes) > 0

    # --- final decision ---
    if has_boxes:
        pred_fracture = True
    elif cls_label == "fracture" and cls_conf > 0.7:
        pred_fracture = True
    else:
        pred_fracture = False

    y_true.append(int(gt_fracture))
    y_pred.append(int(pred_fracture))

    if gt_fracture and pred_fracture:
        tp += 1
    elif not gt_fracture and pred_fracture:
        fp += 1
    elif not gt_fracture and not pred_fracture:
        tn += 1
    elif gt_fracture and not pred_fracture:
        fn += 1

    precision = tp / (tp + fp + 1e-9)
    recall = tp / (tp + fn + 1e-9)
    f1 = 2 * precision * recall / (precision + recall + 1e-9)
    acc = (tp + tn) / (tp + tn + fp + fn + 1e-9)

    precisions.append(precision)
    recalls.append(recall)
    f1s.append(f1)
    accs.append(acc)
    steps.append(idx)

# === final metrics ===
final_precision, final_recall, final_f1, final_acc = precisions[-1], recalls[-1], f1s[-1], accs[-1]
print("\n=== PIPELINE RESULTS (Detector + Classifier) ===")
print(f"Images tested: {len(all_imgs)}")
print(f"TP: {tp}  FP: {fp}  TN: {tn}  FN: {fn}")
print(f"Precision: {final_precision:.3f}")
print(f"Recall:    {final_recall:.3f}")
print(f"F1-score:  {final_f1:.3f}")
print(f"Accuracy:  {final_acc:.3f}")

# === curves ===
def save_curve(x, y, label, name):
    plt.figure()
    plt.plot(x, y, label=label, color="royalblue")
    plt.xlabel("Images processed")
    plt.ylabel(label)
    plt.title(f"{label} curve")
    plt.ylim(0, 1)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.savefig(save_dir / f"{name}.png", dpi=300)
    plt.close()

save_curve(steps, precisions, "Precision", "BoxP_curve")
save_curve(steps, recalls, "Recall", "BoxR_curve")
save_curve(steps, f1s, "F1", "BoxF1_curve")
save_curve(steps, f1s, "Precision–Recall", "BoxPR_curve")  # re-using F1 just for naming symmetry

# === confusion matrix ===
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, y_pred)
cmn = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-9)
labels = ["Healthy", "Fracture"]

for mat, name in [(cm, "confusion_matrix"), (cmn, "confusion_matrix_normalized")]:
    plt.figure()
    sns.heatmap(mat, annot=True, fmt=".2f" if "normalized" in name else "d",
                xticklabels=labels, yticklabels=labels, cmap="Blues")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(name.replace("_", " ").title())
    plt.tight_layout()
    plt.savefig(save_dir / f"{name}.png", dpi=300)
    plt.close()
```
